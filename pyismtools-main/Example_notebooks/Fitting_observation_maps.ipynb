# I - Load the observation datafile

The observed map should be provided as two ascii files: one for the observed values and one for the error levels.
Each of these two files should contain one line per pixel, the first two columns being the x and y coordinate of the pixel, and the other columns giving the value of intensity/error level for the different lines that have been observed. Note that all pixels should contain the same lines. If a line is not detected in a pixel, you can put a NaN value for this line for the corresponding pixel, both for the intensity value and for the error levels (upper limits are not usable in the fit for now).

The header should provide the number of observed lines and the list of names of the lines in the same order as they will be provided for each pixel. For instance, you can find below the beginning of a valid map file for the intensity values :
```
# Nb of observables :
4
# List of observable names :
13CO J=2-1
C+ 158um    
CH+ J=1-0   
CO J=10-9    
## Data array
#pixel_x pixel_y  13CO J=2-1           C+ 158um             CH+ J=1-0            CO J=10-9
0        0        6.43445e-08          nan                  nan                  -1.5698655e-06                     
1        0        6.537173e-08         nan                  nan                  -1.3727921e-06                     
2        0        6.475346e-08         nan                  2.4983203e-07        -2.108265e-07 
3        0        6.151421e-08         nan                  1.8396711e-06        2.2226504e-06  
4        0        6.7370564e-08        nan                  1.6454493e-06        3.3636893e-06 
5        0        8.6569834e-08        nan                  2.1571232e-07        5.522471e-06     
```
The format of the header is the same for the intensity file and for the error file.

````python
# We load an observation map
OMC1_map = ism.Observation_map.from_ASCII_file("../Map_observations/OMC1/OMC1_int.dat","../Map_observations/OMC1/OMC1_err.dat","OMC1 Herschel map")
print(OMC1_map) # displays information about your observation map
````

````python
# we can visualize the observation maps easily:
ism.viz_tools.plot_intensity_maps(OMC1_map)
plt.show()
````

# II - Load the model grid

You have two ways to load a model grid.

The first is to load a model grid from the online database of PDR models ISMDB. 

You first need to find the grid you want to use among the grids listed [here](https://app.ism.obspm.fr/ismdb/). The screenshot below indicates where to find the grid ID:

![title](ISMDB_screenshot.png)

Then, you need to build the list of line names for all the lines you want to include. To find the name of a given line, click the button "Fit models to observations", and in the section "3 - Observational constraints", use the quantity search bar to find the line you are looking for. You will then see the full name of the quantity appear just below the search bar (cf. screenshot below):

![title](quantity_name_Screenshot.png)

````python
grid_ID = "P154G3_P_210723"

observable_list = ["I(13CO J=2->J=1 angle 00 deg)",
                   "I(C+ El=2P,J=3/2->El=2P,J=1/2 angle 00 deg)",
                   "I(CH+ J=1->J=0 angle 00 deg)",
                   "I(CO v=0,J=10->v=0,J=9 angle 00 deg)",
                   "I(CO v=0,J=2->v=0,J=1 angle 00 deg)",
                   "I(HCO+ J=3->J=2 angle 00 deg)"
                  ]

my_grid = ism.ismdb_api.get_grid(grid_ID,observable_list)

print(my_grid) # this displays a summary of your model grid
````

# III - Translate your observable names to have the same names in the observation and in the model grid

````python
# Define a dictionnary for translating line names

correspondance_dict = {
                         "13CO J=2-1":"I(13CO J=2->J=1 angle 00 deg)",
                         "C+ 158um":"I(C+ El=2P,J=3/2->El=2P,J=1/2 angle 00 deg)",
                         "CH+ J=1-0":"I(CH+ J=1->J=0 angle 00 deg)",
                         "CO J=10-9":"I(CO v=0,J=10->v=0,J=9 angle 00 deg)",
                         "CO J=2-1":"I(CO v=0,J=2->v=0,J=1 angle 00 deg)",
                         "HCO+ J=3-2":"I(HCO+ J=3->J=2 angle 00 deg)"
                      }

# Translate the names in the model grid to be the same than in the observation
translation = ism.Translation("my_translation",correspondance_dict)
my_grid.translate_names(translation)
# (WARNING : if you execute this cell several times, it will translate back and forth repeatedly
# and you will end up with the initial line names if you have executed it an even number of times)

print(my_grid) # the line names are now the same as for the observation
````

# IV - Fit the model grid to the observation

When fitting PDR models to observations, it is recommended to account for the fact that the model itself has uncertainties, typically of 30-50%. Otherwise, the fit might try to reproduce at all cost the high SNR lines and completely miss the low SNR lines, while a model fitting both reasonably well might actually exist.

````python
# Create a new observation object in which we include the model uncertainties :

OMC1_map2 = OMC1_map.copy()

# set errors to 50% for all observed lines

OMC1_map2.set_multiplicative_noise_threshold(threshold=0.50)
print(OMC1_map2)
# For all lines where the observational noise was lower than 50% of the observed intensity, 
# the noise was replaced by a multiplicative noise of 50%.
````

We can then fit the model to each pixel of the observed map:

````python
# we first build and interpolation of the model grid 
my_approximator = ism.Grid_approximator.interpolate_from_grid(my_grid)

# and then fit the observation
best_models, best_chi2s = ism.fit_tools.find_best_interpolated_model(OMC1_map,my_approximator,N_repetitions = 10,cost_function="chi2")
# N_repetitions is the number of times the minimization is repeated (from random initial guesses) to avoid 
# bad local minima.
# Here we specify the cost function explicitly, as the default is a reduced chi2, which is not defined 
# when there are less observations than free parameters (which is the case here for some pixels).
````

We can now visualize the value of the $\chi^2$ of the best model at each pixel:

````python
ism.viz_tools.map_fit_quality(OMC1_map,best_chi2s)
plt.show()
````

We can also visualize the map of best parameter values:

````python
ism.viz_tools.map_best_parameters(OMC1_map,best_models)
````
